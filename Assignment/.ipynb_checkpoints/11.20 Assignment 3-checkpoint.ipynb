## 大樂透中獎號碼爬蟲

### 用Request取得網頁碼

```python
from bs4 import BeautifulSoup
import requests

Lotto_HTML = 'https://www.taiwanlottery.com.tw/Lotto/Lotto649/history.aspx' #目標網址
req = requests.get(Lotto_HTML, timeout = 30) #用Get抓取，用Timeout設回應的限制時長
print(req.text) #看抓到的內容
```

### 觀察原始碼，找出開獎號碼的格式，用迴圈設成條件

```python
#蒐集開獎號碼的的關鍵格式
WinNum = ['Lotto649Control_history_dlQuery_No1_','Lotto649Control_history_dlQuery_No2_','Lotto649Control_history_dlQuery_No3_','Lotto649Control_history_dlQuery_No4_','Lotto649Control_history_dlQuery_No5_','Lotto649Control_history_dlQuery_No6_','Lotto649Control_history_dlQuery_SNo_']

def search_num(css_class):
    global WinNum #做全域變數宣告
    if(css_class != None):
        for i in range(len(WinNum)):
            if WinNum [i] in css_class:
                return css_class
```

### 解析原始碼，用find_all篩出中獎號碼

```python
soup = BeautifulSoup(req.text,'lxml') #用lxml解析器原始碼

print("108年11月大樂透號碼：")
for num in soup.find_all(id=search_num):
    print(num.text)
```
