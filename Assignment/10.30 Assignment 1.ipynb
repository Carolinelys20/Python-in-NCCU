## 整理出各校學生總人數

```python
df = pd.read_csv("107_students.csv", index_col = "學校名稱")
```

```python
df.head()
```

```python
#篩選出學士班且日間部的人數
df = df[(df["等級別"] == "B 學士") & (df["日間∕進修別"] == "D 日")]
```

```python
df["人數總和"] = df.loc[:,"一年級男生":"延修生女生"].sum(axis=1)
```

```python
df_schoolnum = df.groupby("學校名稱").sum()
```

```python
df_schoolnum
```

```python
df_sorted = df_schoolnum.sort_values(by="人數總和", ascending=False)
```
