```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from scipy.integrate import odeint
```

```python
def twolevel(y_list, x):
    y,z = y_list
    return np.array([z, -y])
```

```python
x = np.linspace(0, np.pi*3, num=100)
```

```python
y0 = [1, 1]
```

```python
result = odeint(twolevel, y0, x)
```

```python
plt.plot(x, result[:, 0], label = "y")
plt.plot(x, result[:, 1], label = "z")
plt.legend()
plt.grid()
```
