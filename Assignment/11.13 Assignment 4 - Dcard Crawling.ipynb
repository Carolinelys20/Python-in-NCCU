## Dcard爬蟲 列印出所有板塊名稱

```python
import requests
from bs4 import BeautifulSoup
import re
```

```python
#設定取得的網址：dcard首頁
url = "https://www.dcard.tw/f"
```

```python
#用requests套件設定回傳
resp = requests.get(url)
```

```python
#確認上述回傳是否正常
print(resp.status_code)
```

```python
#用BeautifulSoup解析HTML程式碼
soup = BeautifulSoup(resp.text, "html.parser")
```

```python
#輸出排版過後的HTML碼
print(soup.prettify())
```

```python
#列印網頁標題標籤
title_tag = soup.title
print(title_tag)
```

```python
#萃取標題文字：用string
print(title_tag.string)
```

```python
#用"find_all"找出所有特定的 HTML 標籤節點(找出所有版的名稱)

dcard_subtitle = soup.find_all('a', "ForumEntry_link_3tNC45 ForumEntry_unRead_1ptlOL")
```

```python
#查看所有Dcard板塊subtitle
dcard_subtitle
```

```python
print("Dcard所有板塊名稱：")
print("\n")
for subtitle in soup.find_all('a', "ForumEntry_link_3tNC45 ForumEntry_unRead_1ptlOL"):
    print(subtitle.text)
```
