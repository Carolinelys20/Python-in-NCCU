```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## PCA

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
x = iris.data
y = iris.target
```

```python
from sklearn.decomposition import PCA
```

## Step1. 開PCA機

#### 選一個最佳平面把所有特徵投影上去

```python
#n_components 把數據壓到""維
pca = PCA(n_components = 2)
```

## Step2. 訓練

```python
pca.fit(x)
```

## Step3. 預測

```python
#原先x的87號
x[87]
```

```python
#用PCA的「預測」將數據轉成2維
X = pca.transform(x)
```

```python
#更新後X的87號
X[87]
```

```python
#畫出PCA的分群結果
plt.scatter(X[:,0], X[:,1], c=y)
```

## 準備訓練/測試資料

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y,
                                                   test_size=0.2,
                                                   random_state=0)
```

## Step1. 開SVC函數學習機

```python
from sklearn.svm import SVC
```

```python
clf = SVC(gamma="auto")
```

## Step2. 訓練

```python
clf.fit(x_train, y_train)
```

```python
y_predict = clf.predict(x_test)
```

```python
y_predict
```

```python
x0 = np.arange(-4, 4.2, 0.02)
y0 = np.arange(-1.5, 1.7, 0.02)

xm, ym = np.meshgrid(x0, y0)
P = np.c_[xm.ravel(), ym.ravel()]
z = clf.predict(P)
Z = z.reshape(xm.shape)
plt.contourf(xm, ym, Z, alpha=0.3, cmap="Paired")
plt.scatter(x_test[:,0], x_test[:,1], c=y_test, cmap="Paired")
```

## K-Means 自動分類

```python
x = np.random.rand(100,2)
```

```python
plt.scatter(x[:,0], x[:,1])
```

## Step.1 打開K-Means 學習機

```python
from sklearn.cluster import KMeans
```

```python
clf = KMeans(n_clusters=3)
```

## Step2. 學習

```python
clf.fit(x)
```

## Step3. 預測

```python
clf.predict(x)
```

```python
clf.labels_
```

```python
#畫圖看電腦如何分成三類
plt.scatter(x[:,0], x[:,1], c=clf.labels_)
```

```python
x0 = y0 =np.arange(-0.2, 1.2, 0.02)
xm, ym = np.meshgrid(x0, y0)

P = np.c_[xm.ravel(), ym.ravel()]
z = clf.predict(P)

Z = z.reshape(xm.shape)
plt.contourf(xm, ym, Z, alpha=0.3, cmap="Paired")
plt.scatter(x[:,0], x[:,1], c=clf.labels_, cmap="Paired")
```

## "make_blobs" 生更像真實世界的數據

```python
from sklearn.datasets.samples_generator import make_blobs
```

```python
#生成500筆資料，有3類，2個變數
x, y = make_blobs(n_samples=500, centers=3,
                 n_features=2,
                 random_state=0)
```

```python
plt.scatter(x[:,0], x[:,1], c=y, cmap="Paired")
```

```python
plt.scatter(x[:,0], x[:,1])
```

## Step1. 打開K-Means 函數學習機

```python
clf = KMeans(n_clusters=3)
```

## Step2. 訓練

```python
clf.fit(x)
```

## Step3. 預測

```python
plt.scatter(x[:,0], x[:,1], c=clf.labels_, cmap="Paired")
```

## Mean-Shifts 完全自動分類

## Step1. 打開函數學習機

```python
from sklearn.cluster import MeanShift
```

```python
clf = MeanShift(bandwidth=0.05)
```

## Step2. 訓練

```python
clf.fit(x)
```

## Step3. 預測

```python
plt.scatter(x[:,0], x[:,1], c=clf.labels_, cmap="Paired")
```

```python
x0 = np.arange(-3.8, 5.2, 0.02)
y0 = np.arange(-1.8, 7, 0.02)

xm, ym = np.meshgrid(x0, y0)

P =np.c_[xm.ravel(), ym.ravel()]
z = clf.predict(P)
Z = z.reshape(xm.shape)

plt.contourf(xm, ym, Z, alpha=0.3, cmap="Paired")
plt.scatter(x[:,0], x[:,1], c=clf.labels_, cmap="Paired")
```
