## map 

```python
items = [500, 40, 2000]
c = 30.8594
```

```python
##用迴圈寫函數(換算)
for i in items:
    p = i*c
    egg.append(p)
```

```python
##用map寫函數(換算)
def exchange(p):
    return p*c
```

```python
exchange(500)
```

```python
egg = list(map(exchange, items))
```

```python
egg
```

```python
## 'lambda' 救急函數

map(lambda p:p*c, items)
```

```python
##上一行map文件要用'list'去看

list(map(lambda p:p*c, items))
```

```python
#把items內的p拿出來做p*c

[p*c for p in items]
```

## filter 

```python
grades = [50, 59, 65, 80, 100, 35, 40, 88]
```

```python
#寫函數去做正確與否判斷

def spam(g):
    return g>=60
```

```python
spam(50)
```

```python
spam(80)
```

```python
# 用filter去判斷spam內的grades

list(filter(spam, grades))
```

```python
# 用lambda去簡寫(免定義spam函數)

list(filter(lambda g:g>=60, grades))
```

## print 的格式化

```python
name = '人'
age = 25
```

```python
print('您好，我是' + name + ',今年' + str(age) + '歲。')
```

```python
### method 1
###用%s指定字串/%d指定數字 位置，在後方用%()指定放的內容

egg = "您好，我是%s，今年%d歲。"%(name, age)
print(egg)
```

```python
### method 2: format
###使用{}標明位置，用format寫出內容

egg = '您好，我是{}，今年{}歲。'.format(name, age)
print(egg)
```

```python
### method 3: f-string

egg = f"您好，我是{name}，今年{age}歲。"
print(egg)
```

## 數字照我們的意思顯示

```python
c = 30.8594353
```

```python
print(f"1美金和台幣{c}元。")
```

```python
##使用c:.2f取到小數點後兩位(f=float浮點數)

print(f"1美金和台幣{c:.2f}元。")
```

```python
##c:10.2f(10個空格並取小數點後兩位)
print(f"1美金和台幣{c:10.2f}元。")
```

## 更完整的print控制

```python
egg = "╭∩╮（︶︿︶）╭∩╮"

spam =f"⚇{egg}⚇"

print(spam)
```

```python
## round 1: 空20格

spam = f"⚇{egg:20s}⚇"

print(spam)
```

```python
## round 2: 空20格(往右靠)

spam = f"⚇{egg:>20s}⚇"

print(spam)
```

```python
## 畫聖誕樹

level = 10

for n in range(level):
    s = "*"*(2*n +1)
    print(f"{s:^21s}")
```

```python
spam = f"⚇{egg:*^20s}⚇"

print(spam)
```

```python
#數字前補0

##方法一：用0>3s，表示靠右三格用0補空
n = 21

spam = f"{str(n):0>3s}"
print(spam)
```

```python
##方法二：前面加上字串"000"，再取後三位

n = 87

spam = ("000" + str(n))[-3:]

print(spam)
```

```python
##方法三： 使用zfill

n = str(87)

spam = n.zfill(3)
print(spam)
```
